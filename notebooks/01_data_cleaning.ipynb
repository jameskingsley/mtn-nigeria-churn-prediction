Loadind Libraries

```python
import os
import pandas as pd
import numpy as np
```

Loading data

```python
# 1. Define paths dynamically
RAW_DATA_PATH = os.path.join("..", "data", "raw", "mtn_customer_churn.csv")
PROCESSED_DATA_PATH = os.path.join("..", "data", "processed", "mtn_customer_churn_cleaned.csv")

print("Loading raw MTN Nigeria customer churn dataset...")
df = pd.read_csv(RAW_DATA_PATH)
```

```python
df.head()
```

```python
df.info()
```

Standardize Column Names

```python
# Standardizing columns to snake_case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
print("New Column Names:\n", df.columns.tolist())
```

Fix Datetime Formatting

```python
# Converting to formal datetime
df['date_of_purchase'] = pd.to_datetime(df['date_of_purchase'], format='%b-%y', errors='coerce')
print("Datetime check:\n", df['date_of_purchase'].head())
```

Handle Target-Dependent Missing Values

```python
# Impute missing churn reasons for active users
df['reasons_for_churn'] = df['reasons_for_churn'].fillna('not_churned')

# Verifying no more missing values exist across the dataset
print("Missing values remaining:", df.isnull().sum().sum())
```

Normalizing and Clean Categorical Values

```python
categorical_cols = ['state', 'mtn_device', 'gender', 'customer_review', 'subscription_plan', 'customer_churn_status']

for col in categorical_cols:
    df[col] = df[col].astype(str).str.strip().str.lower()

print("Unique targets:", df['customer_churn_status'].unique())
```

High-Fidelity Data Validation 

```python
# Recalculating and verifying revenue integrity
calculated_revenue = df['unit_price'] * df['number_of_times_purchased']
mismatches = (df['total_revenue'] != calculated_revenue).sum()

print(f"Number of revenue mismatch records: {mismatches}")
```

Export Cleaned File to the Processed Directory

```python
import os

# Create directory if it doesn't exist
os.makedirs("../data/processed", exist_ok=True)

# Save file
PROCESSED_DATA_PATH = os.path.join("..", "data", "processed", "mtn_customer_churn_cleaned.csv")
df.to_csv(PROCESSED_DATA_PATH, index=False)

print(f"Success! Cleaned dataset successfully stored at: {PROCESSED_DATA_PATH}")
```
